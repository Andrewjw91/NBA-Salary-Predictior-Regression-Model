```python
from __future__ import print_function, division
import requests
import pandas as pd
import numpy as np
import os.path
from bs4 import BeautifulSoup

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")

%matplotlib inline

from sklearn.linear_model import LinearRegression, Lasso, LassoCV
import statsmodels.formula.api as smf
```

```python
data = pd.read_csv(r'C:\Users\andro\Metis\module2_repo\NBA\bballref.csv')
df = pd.DataFrame(data, columns=['Player', 'Pos', 'Age', 'Tm', 'G', 'GS', 'MP', 'FG', 'FGA', 'FG%', '3P',
       '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT', 'FTA', 'FT%', 'ORB',
       'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS', 'YEAR'])
```

```python
data2 = pd.read_csv(r'C:\Users\andro\Metis\module2_repo\NBA\salaries.csv')
df2 = pd.DataFrame(data2, columns=['NAME','SALARY','YEAR'])
df2 = df2.rename(columns={'NAME':'Player'})
df2.head()
```

```python
combined = pd.merge(df, df2, on=['Player','YEAR'])
combined.head()
```

```python
combined['SALARY'] = combined['SALARY'].astype("string")
combined['SALARY'] = combined['SALARY'].str.replace('$', '').str.replace(',', '')
combined['SALARY'] = combined['SALARY'].astype('int64')
```

```python
combined['3P%'] = combined['3P%'].fillna(0.0)
combined['FT%'] = combined['FT%'].fillna(0.0)
```

```python
combined.to_csv('combined.csv')
```

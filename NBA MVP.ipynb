# MVP Linear Regression Model Predicting NBA Salaries Based on Player Performance Statistics

I was able to create a Linear Regression model to predict NBA Player Salaries based on Player performance statistics. The performance statistics I chose for mulitvariate feature selection were: ('Age', 'MP', 'FG%', '3P%', '2P%', 'FT%', 'TRB', 'AST', 'BLK', 'PTS','YEAR')

My Linear Regression model had an R2 value = 0.533 on the training data. When applied to my test data the R2 value was 0.566. So far the feature stats I select produced the highest R2 value on my test data.
